### Chi-square for categorical attributes

```python
import numpy as np
import scipy.stats as stats
from scipy.stats import chi2_contingency

cat_df = pd.DataFrame(data, columns=['Status of existing checking account',
        'Credit history',
        'Purpose',
        'Savings account/bonds',
        'Present employment since',
        'Personal status and sex',
        'Other debtors / guarantors',
        'Present residence since',
        'Property',
        'Other installment plans',
        'Housing',
        'Job',
        'Telephone',
        'foreign worker',
        'score'])

features = ['Status of existing checking account',
        'Credit history',
        'Purpose',
        'Savings account/bonds',
        'Present employment since',
        'Personal status and sex',
        'Other debtors / guarantors',
        'Present residence since',
        'Property',
        'Other installment plans',
        'Housing',
        'Job',
        'Telephone',
        'foreign worker']

# The function 'chi2_contingency' computes the chi-square statistic and p-value for the hypothesis test of 
# independence of the observed frequencies in the contingency table observed. The expected frequencies are 
# computed based on the marginal sums under the assumption of independence
for f in features:
    X= cat_df[f].astype(str)
    Y= cat_df["score"].astype(str)
    df_observed = pd.crosstab(Y,X)
    chi2, pval, dof, expected = stats.chi2_contingency(df_observed.values)

    if pval<0.05:
        print("{0} is IMPORTANT \n".format(f))
    else:
        print("{0} is NOT IMPORTANT \n".format(f))
```

#### Note: After analysing the data contained in the feature 'Present resisdence since' we can safely consider it as a categorical attribute

### Principle Component analysis 

```python
from sklearn.decomposition import PCA

pca_df = pd.DataFrame(data, columns=['Duration in month', 'Credit amount', 'Installment rate of disposable income',
                                              'Age','Number of existing credits',
                                              'Number of people being liable'])

Z_score = (pca_df.to_numpy()-pca_df.to_numpy().mean(axis=0))/pca_df.to_numpy().std(axis=0)

pca = PCA().fit(Z_score)
                                          
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_)
plt.title('Principal Component Analysis')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

#### From the graph we can assert that the first 5 components contain 95% of information and hence the last component can be ignored in order to improve the performance at the cost of some accuracy
